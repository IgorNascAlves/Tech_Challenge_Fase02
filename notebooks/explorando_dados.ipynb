```python
import pandas as pd
```

```python
path = '../data/Dados Históricos - Ibovespa.csv'
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
#transformar a coluna data em datetime dia?mes?ano
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')
```

```python
#print data mais antiga e mais recente
print("Data mais antiga: ", df['Data'].min())
print("Data mais recente: ", df['Data'].max())
```

```python
#criar um grafico com o df utilizando o seaborn Data X Ùltimo
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')



plt.figure(figsize=(16,8))
sns.lineplot(x='Data', y='Último', data=df)
#coloque um titulo no grafico
plt.title('Ibovespa')
#coloque um label no eixo x
plt.xlabel('Data')
#coloque um label no eixo y
plt.ylabel('Último')
plt.xticks(rotation=45)
plt.show()
```

```python
#criar um grafico com o df utilizando o seaborn Data X Ùltimo, Maxina, Minima, Abertura
plt.figure(figsize=(16,8))
sns.lineplot(x='Data', y='Último', data=df, label='Último')
sns.lineplot(x='Data', y='Máxima', data=df, label='Máxima')
sns.lineplot(x='Data', y='Mínima', data=df, label='Mínima')
sns.lineplot(x='Data', y='Abertura', data=df, label='Abertura')
#coloque um titulo no grafico
plt.title('Ibovespa')
#coloque um label no eixo x
plt.xlabel('Data')
#coloque um label no eixo y
plt.ylabel('Último')
plt.xticks(rotation=45)
plt.show()
```
