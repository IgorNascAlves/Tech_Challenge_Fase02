```python
import pandas as pd
```

```python
path = '../data/Dados Históricos - Ibovespa.csv'
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
#transformar a coluna data em datetime dia?mes?ano
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')
```

```python
#criar um grafico com o df utilizando o seaborn Data X Ùltimo
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')



plt.figure(figsize=(16,8))
sns.lineplot(x='Data', y='Último', data=df)
#coloque um titulo no grafico
plt.title('Ibovespa')
#coloque um label no eixo x
plt.xlabel('Data')
#coloque um label no eixo y
plt.ylabel('Último')
plt.xticks(rotation=45)
plt.show()
```
